## Un poco de teorÃ­a... **tipo de datos RANGE** 

*   Un rango se define como una lista inalterable de nÃºmeros enteros en sucesiÃ³n aritmÃ©tica
*   Un range se define creando el tipo de datos con hasta tres valores numÃ©ricos ->	
m: el valor inicial | 
n: el valor final (que no se alcanza nunca) | 
p: el paso (la cantidad que se avanza cada vez).



## Ejemplo del tipo de datos RANGE

```
#range(n) con un argumento: Se crea una lista de n nÃºmeros enteros consecutivos que empieza en 0 y acaba en n - 1.
valor = range(10)
print(valor)
print(list(valor))

#range(m, n) con dos argumentos: Se crea una lista de enteros consecutivos que empieza en m y acaba en n - 1.
print(list(range(5, 10)))

#range(m, n, p) con tres argumentos: Se crea una lista de enteros consecutivos que empieza en m y acaba justo antes de superar o igualar a n, aumentando los valores de p en p.
print(list(range(10, 0, -2)))
```


````python
#range(n) con un argumento: Se crea una lista de n nÃºmeros enteros consecutivos que empieza en 0 y acaba en n - 1.
valor = range(10)
print(valor)
print(list(valor))

#range(m, n) con dos argumentos: Se crea una lista de enteros consecutivos que empieza en m y acaba en n - 1.
print(list(range(5, 10)))

#range(m, n, p) con tres argumentos: Se crea una lista de enteros consecutivos que empieza en m y acaba justo antes de superar o igualar a n, aumentando los valores de p en p.
print(list(range(10, 0, -2)))
````

## Un poco de teorÃ­a... **tipo de datos LIST** 

*   Una lista es una variable que almacena arrays, por lo que internamente cada posiciÃ³n puede ser un tipo de datos distinto.
*   Los elementos de una lista pueden accederse mediante su Ã­ndice, siendo 0 el Ã­ndice del primer elemento
*   Mediante el acceso a un Ã­ndice o posiciÃ³n concreta, podemos asignar un valor nuevo, por lo que, a diferencia de las tuplas, las listas sÃ­ son tipos de dato mutables.



## Ejemplo del tipo de datos LIST

````python
Frutas = ["Platano", "Pera", "Manzana", "Pera"]
print(Frutas)

Frutas = ["Platano", "Pera", "Manzana", "Pera"]
print(Frutas[0])
print(Frutas[3])

Frutas = ["Platano", "Pera", "Manzana", "Pera"]
print(Frutas)
Frutas[3] = "Kiwi"
print(Frutas)
print("\n")






````

````python
print ("------------MÃTODOS INTEGRADOS--------------")
#EJEMPLOS DE MÃTODOS INTEGRADOS DEL TIPO DE DATOS LIST

print("\n")

print("append")
#append() - Este mÃ©todo agrega un elemento al final de una lista.
Frutas = ["Platano", "Pera", "Manzana", "Pera"]
print(Frutas)
Frutas.append("Kiwi")
print(Frutas)
print("\n")

print("count")
#count() - Este mÃ©todo recibe un elemento como argumento, y cuenta la cantidad de veces que aparece en la lista. De igual manera que actÃºa el mÃ©todocount() en las tuplas.
Frutas = ["Platano", "Pera", "Manzana", "Pera"]
print(Frutas)
print(Frutas.count("Pera"))
print("\n")

print("extend")
#extend() - Este mÃ©todo extiende una lista agregando un iterable al final.
Frutas = ["Platano", "Pera", "Manzana", "Pera"]
print(Frutas)
Frutas.extend("Kiwi")
print(Frutas)
print("\n")

print("index")
#index() - Este mÃ©todo recibe un elemento como argumento, y devuelve el Ã­ndice de su primera apariciÃ³n en la lista
versiones = [1.0, 3.2, 4, 5, 6.5]
print(versiones.index(4))
print("\n")

print("insert")
#insert() - A esta funciÃ³n se le pasan dos argumentos, de la forma insert(i, x) y lo que hace es insertar el elemento x en la posiciÃ³n i. 
versiones = [1.0, 3.2, 4, 5, 6.5]
print(versiones)
versiones.insert(2, -5)
print(versiones)
print("\n")

print("pop")
#pop() - Similar a la funciÃ³n de desapilar si trabajÃ¡semos con pilas, esta funciÃ³n retorna el Ãºltimo elemento de la lista y lo borra. 
versiones = [1.0, 3.2, 4, 5, 6.5]
print(versiones.pop())
print(versiones)
print("\n")

print("remove")
#remove() - A este mÃ©todo se le pasa como argumento un valor y lo que hace es buscar la primera apariciÃ³n de este en la lista y lo borra.
versiones = [1.0, 3.2, 4, 5, 6.5]
print(versiones)
versiones.remove(3.2)
print(versiones)
print("\n")

print("reverse")
#reverse() - Este mÃ©todo invierte el orden de los elementos en la lista; es decir, ordena de forma inversa los elementos en base a su Ã­ndice o posiciÃ³n.
versiones = [1.0, 3.2, 4, 5, 6.5]
print(versiones)
versiones.reverse()
print(versiones)
print("\n")

print("sort")
#sort() - Este mÃ©todo ordena los elementos de una lista.
versiones = [1.0, 3.2, 8, 5, 6.5]
print(versiones)
versiones.sort()
print(versiones)
versiones.sort(reverse=True)
print(versiones)
print("\n")
````

````python
print ("------------MÃTODOS INTEGRADOS--------------")
#EJEMPLOS DE MÃTODOS INTEGRADOS DEL TIPO DE DATOS DICCIONARIO

print("\n")

print("clear")
#clear() - Borra todos los valores del diccionario
dic = dict(
    Nombre="Josep",
    Tel=688777555,
    Material=["PC", "Teclado", "WebCam"]
)
print(dic, type(dic))
dic.clear()
print(dic, type(dic))
print("\n")

print("copy")
#copy() - Replica el objeto diccionario
dic = dict(
    Nombre="Josep",
    Tel=688777555,
    Material=["PC", "Teclado", "WebCam"]
)
dic2 = dict()
print(dic, dic == dic2)
dic2 = dic.copy()
print(dic2, dic == dic2)
print("\n")

print("fromkeys")
#fromkeys() - Esta funciÃ³n nos ayuda a generar un diccionario a partir de las claves proporcionadas de forma sencilla y utilizando un solo mÃ©todo. 
variables = ('Nombre', 'Tel', 'Material')
dic = dict.fromkeys(variables)
print(dic)
dic = dict.fromkeys(variables, "Rellenar")
print(dic)
print("\n")

print("get")
#get() - Nos devuelve el valor en base a la clave que busquemos. 
dic = dict(
    Nombre="Josep",
    Tel=688777555,
    Material=["PC", "Teclado", "WebCam"]
)
print(dic.get("Nombre"))
print("\n")

print("items")
#items() - Este mÃ©todo devuelve una lista de pares de diccionarios (clave, valor), como 2 tuplas.
dic = dict(
    Nombre="Josep",
    Tel=688777555,
    Material=["PC", "Teclado", "WebCam"]
)
print(dic.items())
print("\n")

print("keys")
#keys() - Este mÃ©todo devuelve una lista de las claves del diccionario:
dic = dict(
    Nombre="Josep",
    Tel=688777555,
    Material=["PC", "Teclado", "WebCam"]
)
print(dic.keys())
print("\n")

print("pop")
#pop() - Este mÃ©todo nos muestra el valor de una clave concreta que le especifiquemos como argumento y la borra del diccionario, junto con su valor. Lanza una excepciÃ³n KeyError si la clave no es encontrada.
dic = dict(
    Nombre="Josep",
    Tel=688777555,
    Material=["PC", "Teclado", "WebCam"]
)
print(dic.pop('Nombre'))
print(dic.items())
print("\n")

print("popitem")
#popitem() - Este mÃ©todo muestra la clave y su valor del primer elemento del diccionario y lo borra. Lanza una excepciÃ³n KeyError si el diccionario esta vaciÃ³.
dic = dict(
    Nombre="Josep",
    Tel=688777555,
    Material=["PC", "Teclado", "WebCam"]
)
print(dic.popitem())
print(dic.items())
print(dic.popitem())
print(dic.items())
print(dic.popitem())
print(dic.items())
print("\n")

print("setdefault")
#setdefault() - Este mÃ©todo es similar a get(key, default_value), pero ademÃ¡s asigna la clave key al valor por default_value para la clave si esta no se encuentra en el diccionario.
dic = dict(
    Nombre="Josep",
    Tel=688777555,
    Material=["PC", "Teclado", "WebCam"]
)
nombre = dic.setdefault('Nombre')
print(nombre)
print("\n")

print("update")
#update() - Este mÃ©todo actualiza un diccionario aÃ±adiendo parejas de clave/valor en un diccionario paralelo. 
dic = dict(
    Nombre="Claudio",
    Tel=688777555,
    Material=["PC", "Teclado", "WebCam"]
)
dic_extra = dict(
    Apellidos="Gomez"
)
print(dic)
print(dic_extra)
dic.update(dic_extra)
print(dic)
print("\n")

print("values")
#values() - Este mÃ©todo devuelve una lista de los valores del diccionario:
dic = dict(
    Nombre="Claudio",
    Tel=688777555,
    Material=["PC", "Teclado", "WebCam"]
)
print(dic.values())

````
